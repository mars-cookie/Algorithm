1096 : [기초-2차원배열] 바둑판에 흰 돌 놓기(설명)
1. 2차원 배열 만드는법 [[리스트] 이걸 반복해서 2번째 리스트]
2. print()를 통해 줄 바꾸지 않게 해놓은 것을 제어하기

```
a=int(input())
b=[[0 for j in range(19)] for i in range(19)]

for i in range(a):
  c,d=input().split()
  c=int(c); d=int(d)
  b[c-1][d-1]=1
for i in range(len(b)):            # 세로 크기
    for j in range(len(b[i])):     # 가로 크기
        print(b[i][j], end=' ')
    print()
```

1. 2중 리스트의 열기준 인덱싱은 없음 -> for로 돌려야됌

```
c=[]
for i in range(19):
  c.append(list(map(int,input().split())))
a=int(input())
for i in range(a):
  d,e=input().split()
  d=int(d);e=int(e)
  c[d-1][0:19]=list(map(lambda x:1-x,c[d-1][0:19]))
  for j in range(19):
    c[j][e-1]=1-c[j][e-1]
  

for i in range(19):
  for j in range(19):
    print(c[i][j],end=' ')
  print()
```

1099 : [기초-2차원배열] 성실한 개미

```
# 2차원 배열 입력
a=[]
for i in range(10):
  a.append(list(map(int,input().split())))

# 초기 위치
row=1;col=1
for i in range(100):
  if a[row][col]==2:
    a[row][col]=9
    break
  a[row][col]=9
  # 오른쪽이 0이나 2라면 오른쪽으로 이동
  
  if a[row][col+1] in [0,2]:
    col+=1
    if a[row][col]==2:
      a[row][col]=9
      break
  # (오른쪽,아래)가 (1,0)이면 아래로 이동
  elif (a[row+1][col] in [0,2]):
    row+=1
    if a[row][col]==2:
      a[row][col]=9
      break
  # (오른쪽,아래)가 (1,1)이면 중단
  elif (a[row][col+1]==1 & a[row+1][col]==1):
    break

# 경로를 출력
for i in range(10):
  for j in range(10):
    print(a[i][j],end=' ')
  print()
```
